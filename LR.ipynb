```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train.csv")
```

```python
df.head()
```

```python
x = np.array(df['x'])
x = x.astype('float64')
```

```python
y = np.array(df['y'])
y = x.astype('float64')
```

```python
x,y
```

```python

fig = plt.figure(figsize=(10,10))
plt.scatter(x,y ,alpha=0.5)
```

```python
def gradient_descent(x,y):
    m_curr = 2
    b_curr = 3
    iterations = 10
    n =len(x)
    learning_rate = 0.01
    
    for i in range(iterations):
        y_predicted = m_curr*x + b_curr
        cost = (1/n)*sum([val**2 for val in (y-y_predicted)])
        md = (1/2*n)*sum(x*(y-y_predicted))
        bd = (1/2*n)*sum(y-y_predicted)
        
        m_curr = m_curr - learning_rate*md
        b_curr = b_curr - learning_rate*bd
        
        print(f"slope={m_curr}  intercept={b_curr} cost ={cost} ,iteration ={i}")
        
```

```python
gradient_descent(x,y)
```
